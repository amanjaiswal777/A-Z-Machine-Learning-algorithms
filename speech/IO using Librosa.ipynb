#### sampling rate of the audio file

```python
sr
```

```python
y, sr = librosa.load(librosa.util.example_audio_file(), offset=30, duration=5)
mfcc = librosa.feature.mfcc(y=x, sr=sr, n_mfcc=128)
```

```python
mfcc.shape
```

```python
import librosa.display
```

```python
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfcc, x_axis='time', sr=16000, cmap='Spectral')
plt.colorbar()
plt.title('MFCC')
plt.tight_layout()
```
